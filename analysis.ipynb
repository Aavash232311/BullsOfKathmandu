<b> General Analysis </b>


#### Inspecting the Dataset `company_list`

```python
%%sql

SELECT * FROM company_list
    ORDER BY as_Of DESC
LIMIT 10;
```

### Count of records in the `company_list` table

```python
%%sql
SELECT COUNT(*) AS total_records_in_company_list FROM company_list;
```

### Inspecting the dataset `combined_price_history`

```python
%%sql
SELECT * FROM combined_price_history LIMIT 5;
```

### Count of records in the `combined_price_history` table

```python
%%sql
SELECT COUNT(*) AS total_records_in_company_price_history FROM combined_price_history;
```

### Time frame for `company_list`

```python
%%sql
SELECT MIN(date_of_operation) AS data_from, MAX(date_of_operation) AS data_to FROM company_list;
```

### Time frame for `combined_price_history`

```python
%%sql
SELECT MIN(date_added) AS data_from, MAX(date_added) AS data_to FROM combined_price_history;
```

### Checking if this is unique.

```python
%%sql
SELECT COUNT(*) FROM  combined_price_history WHERE
    company_name = 'Nepal Finance Limited';
```
