### Which sector has the highest `market capitalization`

Side Note: Market capitalization, or market cap, is the total dollar value of a company’s outstanding shares of stock, calculated by multiplying the current share price by the total number of outstanding shares.

```python
%%sql
SELECT
    SUM(market_capitalization_rs) AS net_market_capitalization_rs, sector
    FROM company_list
    GROUP BY sector
ORDER BY net_market_capitalization_rs DESC
```

### `Government Bonds` are not a number, inspecting that little more.

```python
%%sql
SELECT COUNT(*) as government_bound_counts
    FROM company_list
WHERE sector = 'Government Bonds'
```

### Company with the top 20 highest `market capitalization`

```python
%%sql
SELECT SUM(market_capitalization_rs) as net_market_capitalization_rs, company
    FROM company_list
    WHERE market_capitalization_rs IS NOT NULL
    GROUP BY company
ORDER BY net_market_capitalization_rs DESC LIMIT 20;
```

### Sector with the highest `total_paid_up_capital_rs`

Side Note:- Total Paid-Up Capital (Rs) refers to the actual amount of money a company has received from its shareholders in exchange for shares of stock.

```python
%%sql
SELECT
    MAX(total_paid_up_capital_rs) as total_paid_up_capital_rs_sector, sector
    FROM company_list
    WHERE total_paid_up_capital_rs IS NOT NULL
    GROUP BY sector
ORDER BY total_paid_up_capital_rs_sector DESC;
```

### Top 30 company with the highest `total_paid_up_capital_rs`

```python
%%sql
SELECT
    MAX(total_paid_up_capital_rs) as total_paid_up_capital_rs_sector, company, sector
    FROM company_list
    WHERE total_paid_up_capital_rs IS NOT NULL
    GROUP BY company, sector
ORDER BY total_paid_up_capital_rs_sector DESC LIMIT 30;
```
